```python
import numpy as np 
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
cols = ['age', 'sex', 'chest_pain', 'resting_bp', 'cholestoral', 'fbs', 'resting_ecg', 'max_pulse',
'angina', 'oldpeak', 'slope', 'major_vessles', 'thal', 'target']
```

```python
df = pd.read_csv("Heart Disease data.csv",names=cols,header = 0)
```

```python
df.head()
```

```python
for col in df.columns:
    print(col, df[col].unique())
```

```python
df.isnull().sum()
```

# EDA

## Heatmap

```python
plt.figure(figsize=(12, 10))
sns.heatmap(df.corr(), vmin=-1, vmax=1, cmap='YlGnBu',annot=True)
plt.title("Correlation among all the attributes", fontsize = 18)
```

## Pairplot

```python
sns.pairplot(df, hue = 'target')
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data = df, x = 'max_pulse', kde = True, hue = 'target')
plt.xlabel("Maximum Pulse", fontsize=20)
plt.ylabel("Count", fontsize=20)
plt.title("Maximum Pulse wrt target", fontsize=23)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data = df, x = 'cholestoral', kde = True, hue = 'target')
plt.xlabel("Cholestoral", fontsize=20)
plt.ylabel("Count", fontsize=20)
plt.title("Cholestoral wrt target", fontsize=23)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data = df, x = 'resting_ecg', kde = True, hue = 'target')
plt.xlabel("Resting ECG", fontsize=20)
plt.ylabel("Count", fontsize=20)
plt.title("Resting ECG wrt target", fontsize=23)
plt.show()
```

```python

df['target'] = df['target'].astype(str)

plt.figure(figsize=(9, 5))
sns.countplot(x="sex", data=df, hue='target', palette='bright')
plt.xlabel('Gender', fontsize=20)
plt.ylabel('Count', fontsize=20)
plt.title("Gender Count based on heart disease", fontsize=23)
plt.show()
```

```python
plt.figure(figsize=(9, 5))
sns.countplot(x="chest_pain", data = df, hue='target', palette='bright')
plt.xlabel('Chest Pain', fontsize = 20)
plt.ylabel('Count', fontsize = 20)
plt.title("Chest Pain vs Heart Disease", fontsize = 23)
plt.show()
```

```python
plt.figure(figsize=(15,7))
sns.countplot(x="age", data = df, hue='target', palette='bright')
plt.xlabel('Age', fontsize = 20)
plt.ylabel('Count', fontsize = 20)
plt.title("Age Count based on heart disease", fontsize = 23)
plt.show()
```

```python
plt.figure(figsize=(9,5))
sns.lineplot(data = df, x = 'age', y = 'chest_pain', err_style=None, hue = 'sex',palette='bright', ci = 0, legend='full')
plt.xlabel('Age', fontsize = 20)
plt.ylabel('Chest Pain', fontsize = 20)
plt.title("Age vs Chest Pain", fontsize = 23)
plt.show()
```

```python
plt.figure(figsize=(9,5))
sns.lineplot(data = df, x = 'age', y = 'resting_bp', err_style=None, hue = 'sex',palette='bright', ci = 0, legend='full')
plt.xlabel('Age', fontsize = 20)
plt.ylabel('Resting BP', fontsize = 20)
plt.title("Age vs Resting BP", fontsize = 23)
plt.show()
```

```python
plt.figure(figsize=(9,5))
sns.lineplot(data = df, x = 'age', y = 'cholestoral', err_style=None, hue = 'sex',palette='bright', ci = 0, legend='full')
plt.xlabel('Age', fontsize = 20)
plt.ylabel('Cholestoral', fontsize = 20)
plt.title("Age vs Cholestoral", fontsize = 23)
plt.show()
```

```python
plt.figure(figsize=(16,8))
sns.lineplot(data = df, y = 'max_pulse', x = 'slope',hue = 'target',palette='bright', ci = 65)
plt.xlabel('ST Slope', fontsize = 20)
plt.ylabel('Max Pulse', fontsize = 20)
plt.title("ST Slope vs Max Pulse", fontsize = 23)
```

```python
#'resting_bp', 'cholestoral', 'max_pulse'
plt.figure(figsize=(14, 6))
sns.relplot(data=df, x='resting_bp', y='max_pulse', hue='target', size='cholestoral',sizes=(20, 300), alpha=.5, palette="muted", height=6)
plt.xlabel("Resting BP", fontsize = 20)
plt.ylabel("Max Pulse", fontsize = 20)
plt.title("Max Pulse vs Resting BP", fontsize = 23)
```

```python
plt.figure(figsize=(14, 6))
sns.violinplot(data=df, x='thal', y='resting_bp', hue='target', palette="muted")
plt.xlabel("Thalessemia", fontsize = 20)
plt.ylabel("Resting BP", fontsize = 20)
plt.title("Thalessemia vs Resting BP", fontsize = 23)
```
